```python
from flask import Flask, jsonify, request

app = Flask(__name__)


students = {
    1: {"name": "Rohan", "course": "BCA", "age": 20},
    2: {"name": "Priya", "course": "BBA", "age": 21}
}


@app.route("/")
def home():
    return {"message": "Flask Student API Running Successfully"}


@app.route("/students", methods=["GET"])
def get_all_students():
    return jsonify(students)


@app.route("/students/<int:sid>", methods=["GET"])
def get_student_by_id(sid):
    student = students.get(sid)
    if student:
        return jsonify(student)
    return {"error": "Student not found"}, 404

@app.route("/students", methods=["POST"])
def add_student():
    data = request.json
    new_id = max(students.keys()) + 1 if students else 1
    students[new_id] = data
    return {"message": "Student added successfully", "id": new_id}


@app.route("/students/<int:sid>", methods=["PUT"])
def update_student(sid):
    if sid in students:
        students[sid] = request.json
        return {"message": "Student updated successfully"}
    return {"error": "Student not found"}, 404

@app.route("/students/<int:sid>", methods=["DELETE"])
def delete_student(sid):
    if sid in students:
        del students[sid]
        return {"message": "Student deleted successfully"}
    return {"error": "Student not found"}, 404


if __name__ == "__main__":
    app.run(debug=True)
```
