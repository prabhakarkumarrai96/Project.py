```python
class MovieSystem:
    def _init_(self):
        self.movies = {
            1: {"name": "Interstellar", "price": 250, "seats": 50},
            2: {"name": "KGF Chapter 2", "price": 200, "seats": 40},
            3: {"name": "Jawan", "price": 180, "seats": 35},
            4: {"name": "Spider-Man: No Way Home", "price": 300, "seats": 60},
        }

    def show_movies(self):
        print("\nAVAILABLE MOVIES")
        print("------------------------------------")
        for mid, details in self.movies.items():
            print(f"{mid}. {details['name']} | Price: {details['price']} | Seats Left: {details['seats']}")
        print("------------------------------------")

    def book_ticket(self, movie_id, seat_count):
        if movie_id not in self.movies:
            print("Invalid Movie ID.")
            return False

        movie = self.movies[movie_id]

        if seat_count <= 0:
            print("Seat count must be at least 1.")
            return False

        if seat_count > movie["seats"]:
            print("Not enough seats available.")
            return False

        movie["seats"] -= seat_count
        total = seat_count * movie["price"]

        print("\nBOOKING CONFIRMED")
        print("------------------------------------")
        print(f"Movie : {movie['name']}")
        print(f"Seats : {seat_count}")
        print(f"Amount: {total}")
        print("------------------------------------")
        return True

```

```python
system = MovieSystem()

while True:
    print("\nMOVIE TICKET BOOKING SYSTEM")
    print("1. Show Movies")
    print("2. Book Ticket")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        system.show_movies()

    elif choice == "2":
        try:
            mid = int(input("Enter Movie ID: "))
            seats = int(input("Enter number of seats: "))
            system.book_ticket(mid, seats)
        except ValueError:
            print("Please enter only numbers.")

    elif choice == "3":
        print("Exiting system...")
        break

    else:
        print("Invalid option. Try again.")
```
