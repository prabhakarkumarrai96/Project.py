```python
import socket

HOST = "0.0.0.0"
PORT = 5000

server = socket.socket()
server.bind((HOST, PORT))
server.listen(1)

print("ðŸ“¥ Waiting for file...")

conn, addr = server.accept()
print("Connected:", addr)

filename = conn.recv(1024).decode()
f = open(filename, "wb")

while True:
    data = conn.recv(1024)
    if not data:
        break
    f.write(data)

f.close()
conn.close()
server.close()

print(" File received successfully!")
```

```python
import socket

SERVER_IP = "127.0.0.1"   
PORT = 5000
FILE = "test.txt"         

client = socket.socket()
client.connect((SERVER_IP, PORT))

client.send(FILE.encode())

with open(FILE, "rb") as f:
    chunk = f.read(1024)
    while chunk:
        client.send(chunk)
        chunk = f.read(1024)

client.close()
print(" File sent successfully!")
```
